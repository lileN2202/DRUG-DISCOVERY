```python
bioactivity_class = []
for i in selected.Standard_Value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  else:
    bioactivity_class.append("active")
bioactivity_class
```

```python
selected["bioactivity"] = bioactivity_class
selected
```

```python
encoded_class = []
for i in selected.bioactivity:
  if i == "inactive":
    encoded_class.append(0)
  else:
    encoded_class.append(1)
```

```python
selected["numerical_class"] = encoded_class
selected
```

```python
df = selected[["Molecule_ChEMBL_ID", "Smiles"]]
df.to_csv("cleaned_data.csv", index=False)
```

```python
from rdkit.Chem import AllChem
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.ML.Descriptors import MoleculeDescriptors
```

```python
def morgan_fpts(data):
    Morgan_fpts = []
    for i in data:
        mol = Chem.MolFromSmiles(i) 
        fpts =  AllChem.GetMorganFingerprintAsBitVect(mol,2,2048)
        mfpts = np.array(fpts)
        Morgan_fpts.append(mfpts)   
    return np.array(Morgan_fpts)

Morgan_fpts = morgan_fpts(df["Smiles"])
Morgan_fingerprints = pd.DataFrame(Morgan_fpts,columns=['Col_{}'.format(i) for i in range(Morgan_fpts.shape[1])])
Morgan_fingerprints
```

```python
Morgan_fingerprints.to_csv("data_fingerprints.csv", index=False)
```

```python
fingerprints = pd.read_csv("data_fingerprints.csv")
fingerprints
```

```python
fingerprints["bioactivity"] = selected["numerical_class"]
whole_data = fingerprints
whole_data
```

```python
whole_data.to_csv("whole_data.csv", index=False)
```

```python
whole_data= whole_data.dropna()
```

```python
X = whole_data.drop("bioactivity", axis = 1)
y = whole_data["bioactivity"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 39)
```

##   K NEIGHBORS CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
neigh = KNeighborsClassifier(n_neighbors=2,leaf_size=10,weights='distance',algorithm='brute')
```

```python
neigh.fit(X, y)
```

```python
pred=neigh.predict(X_test)
```

```python
proba= neigh.predict_proba(X_test)
proba
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,pred)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,pred)
```

## CROSS VALIDATION

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.svm import SVC
```

```python
svm_model = SVC(C=0.1, gamma='scale', kernel='rbf')
```

```python
k = 3
```

```python
k_fold = KFold(n_splits=k,random_state=22,shuffle=True)
```

```python
cross_val = cross_val_score(svm_model, X, y, cv=k_fold)
```

```python
for fold, score in enumerate(cross_val):
    print(f"Fold {fold+1} accuracy: {cross_val}")
```

```python
mean_accuracy = cross_val.mean()
print(f"Mean accuracy: {mean_accuracy}")
```
